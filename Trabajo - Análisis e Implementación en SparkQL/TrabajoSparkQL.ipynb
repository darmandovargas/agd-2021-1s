```python
# Agrege acá el código para importar las librerias

# La libreria para "encontrar el sevicio" de Spark
import findspark
findspark.init()

# Librerias para "gestionar el servicio" de Spark
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext, SparkSession

# Creamos una aplicación Spark en el Servicio
# Tenga cuidado con las tildes o caracteres especiales en el nombre de la app
AppSpark = SparkConf().setAppName("Evaluacion iSofware")

# definimos un espacio o contexto para la App
ContextoSpark=SparkContext(conf=AppSpark)

# inicio una sesión en el espacio de la App
SesionSpark = SparkSession(ContextoSpark)

# inicio del espacio o contexto SQL
ContextoSql = SQLContext(sparkContext=ContextoSpark, sparkSession=SesionSpark)
```

```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...
ContextoSql.sql("""
CREATE TABLE
asistencia (
        fecha TIMESTAMP,
        correo STRING,
        equipo_pertenece STRING,
        equipo_expone STRING
    )
USING CSV OPTIONS (
        header = 'true',
        multiLine = 'true',
        path = 'asistencia.csv',
        nullvalue = 'NA',
        timestampFormat = "yyyy/MM/dd HH:mm:ss"
    )
""");

ContextoSql.sql("""
CREATE TABLE
evaluacion (
        fecha TIMESTAMP,
        correo STRING,
        equipo_eval STRING,
        intro FLOAT,
        equipo FLOAT,
        problema FLOAT,
        ventajas FLOAT,
        solucion FLOAT,
        producto FLOAT,
        traccion FLOAT,
        mercado FLOAT,
        competencia FLOAT,
        modelo FLOAT,
        inversion FLOAT,
        contacto FLOAT,
        expo1 FLOAT,
        expo2 FLOAT,
        expo3 FLOAT,
        disp_inv STRING,
        observacion STRING
    )
USING CSV OPTIONS (  
        multiLine = 'true',
        path = 'evaluacion.csv',
        nullvalue = 'NA',
        timestampFormat = "yyyy/MM/dd HH:mm:ss",
        header = 'true'
    )
""");
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...
consulta = ContextoSql.sql("SELECT distinct(correo) AS Correo_Estudiante FROM asistencia ORDER BY Correo_Estudiante")
consulta.show()
```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...
consulta = ContextoSql.sql("SELECT correo AS Correo_Estudiante, count(correo) AS Conteo_Asistencias, min(fecha) AS Fecha_hora_primera_asistencia, max(fecha) AS Fecha_hora_ultima_asistencia FROM asistencia GROUP BY Correo_Estudiante ORDER BY Conteo_Asistencias DESC")
consulta.show()
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...
#consulta1 = ContextoSql.sql("SELECT correo, count(correo) AS conteo FROM asistencia WHERE conteo < 3 GROUP BY correo") 
consulta4 = ContextoSql.sql("SELECT correo AS Correo_Estudiante, count(*) AS Conteo_Asistencias FROM asistencia GROUP BY Correo_Estudiante ORDER BY Conteo_Asistencias")
consulta5 = consulta4.select(['Correo_Estudiante','Conteo_Asistencias']).filter(consulta4['Conteo_Asistencias']<3)
consulta5.show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...
consulta5 = ContextoSql.sql("SELECT E.correo AS Correo_Estudiante FROM asistencia A RIGHT JOIN evaluacion E ON A.correo = E.correo WHERE A.correo IS NULL")
consulta5.show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...
# consulta6 = ContextoSql.sql("SELECT equipo_pertenece, distinct(correo) FROM asistencia GROUP BY equipo_pertenece")
# consulta6.show()
asitencia6 = ContextoSql.sql("SELECT distinct(correo) AS Correo_Estudiante, equipo_pertenece FROM asistencia ORDER BY equipo_pertenece")
grupos = asitencia6.select(['equipo_pertenece','Correo_Estudiante'])
grupos.show()
```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...
#consulta7 = ContextoSql.sql("SELECT equipo_pertenece, correo FROM asistencia WHERE SUBSTRING('equipo_pertenece', 15, 80) != SUBSTRING('equipo_expone', 15, 80) GROUP BY equipo_pertenece")
# consulta7 = ContextoSql.sql("SELECT equipo_pertenece, correo FROM asistencia WHERE SUBSTRING('equipo_pertenece', 15, 80) != SUBSTRING('equipo_expone', 15, 80) ORDER BY equipo_pertenece")
# consulta7.show(10)
consulta7 = ContextoSql.sql("SELECT SUBSTRING(equipo_expone,8,100) AS Equipo_Expone, count(SUBSTRING(equipo_expone,8,100)) AS Cantidad_Asitencias from asistencia where SUBSTRING(equipo_expone,8,100)<>SUBSTRING(equipo_pertenece,14,100) group by SUBSTRING(equipo_expone,8,100) ORDER BY Cantidad_Asitencias DESC")
consulta7.show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...
asistencias5 = ContextoSql.sql("SELECT correo AS Correo_Estudiante,SUBSTRING(equipo_pertenece,14,100) as equipo_pertenece, SUBSTRING(equipo_expone,8,100) as equipo_expone  from asistencia where SUBSTRING(equipo_expone,8,100)=SUBSTRING(equipo_pertenece,14,100)")
asistencias5.show()
```

```python
# 9. Consultar la nota promedio por cada ítem (1 al 15), y la nota promedio total del cada equipo. Recuerde que no
# son válidas las evaluaciones realizadas por los miembros del mismo equipo.
# Tu código a continuación...
email_equipos = ContextoSql.table('asistencia').select('correo', 'equipo_pertenece').distinct()

items = [
    'intro',
    'equipo',
    'problema',
    'ventajas',
    'solucion',
    'producto',
    'traccion',
    'mercado',
    'competencia',
    'modelo',
    'inversion',
    'contacto',
    'expo1',
    'expo2',
    'expo3']
     
tabla_evaluacion = ContextoSql.table('evaluacion').select(['correo', 'equipo_eval'] + items).join(email_equipos, on=['correo'], how='left')

tabla_evaluacion_limpia = tabla_evaluacion.filter(tabla_evaluacion['equipo_eval'] != tabla_evaluacion['equipo_pertenece'].substr(14, 100))

promedio_items = tabla_evaluacion.groupBy('equipo_eval').agg({ item: 'avg' for item in items })

notas_promedio = promedio_items.withColumn('nota_promedio_total', sum(promedio_items['avg(' + str(col) + ')'] for col in items) / 15)

notas_promedio.show(n=10, truncate=False, vertical=True)
```
